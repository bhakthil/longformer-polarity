```python
news_df.drop_duplicates(subset=['url'], inplace = True)
print('Unique sources:',len(news_df['source'].unique()))
print('Unique urls:',len(news_df['url'].unique()))
```

```python
news_df.shape
```

```python
reliability_df = pd.read_csv('labels.tab', sep='\t')
reliability_df.columns=['source','reliability']
reliability_df.info()
```

```python
labeled_df = pd.merge(left=news_df, right=reliability_df, on='source', how='left')
labeled_df = labeled_df[labeled_df.reliability.notnull()]
labeled_df.info()
```

```python
labeled_df.head()
```

```python
conditions = [
    (labeled_df['reliability'] == 2.00),
    (labeled_df['reliability'] == 1.00),
    (labeled_df['reliability'] == 0.00)
    ]

# create a list of the values we want to assign for each condition
values = ['Unreliable', 'Mixed', 'Reliable']

# create a new column and use np.select to assign values to it using our lists as arguments
labeled_df['reliability'] = np.select(conditions, values)

# display updated DataFrame
labeled_df.head()
```

```python
labeled_df['source'].value_counts()
```

```python
#filter out value count less than 100
labeled_df = labeled_df[labeled_df['source'].map(labeled_df['source'].value_counts()) > 1000]
```

```python
print(labeled_df['reliability'].value_counts(normalize=True))
print(labeled_df.shape)
```

```python
import urlexpander

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
}

print('Resolve URLs....')
resolved_links = urlexpander.expand(labeled_df['url'], 
                                    chunksize=1024,
                                    n_workers=256, 
                                    cache_file='./temp.json', 
                                    verbose=1, headers=headers, timeout=(7, 20))
labeled_df['resolved_url'] = resolved_links

indices = labeled_df[labeled_df.resolved_url.str.contains("__CLIENT_ERROR__") | 
              labeled_df.resolved_url.str.contains("__CONNECTIONPOOL_ERROR__")].index
labeled_df = labeled_df.drop(indices)
labeled_df.shape
```

```python
import tldextract
def extract_domain(url):
    ext = tldextract.extract(url)
    if(ext.subdomain=='www' or ext.subdomain==''):
        return '.'.join([ext.domain, ext.suffix])
    else:
        return '.'.join([ext.subdomain, ext.domain, ext.suffix])
```

```python

print('Resolving domain....')
labeled_df['article_domain'] = labeled_df['resolved_url'].apply(lambda x:extract_domain(x))
labeled_df.shape
```
